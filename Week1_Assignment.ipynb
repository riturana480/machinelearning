```python
#Task 1
data = [12, 15, 14, 10, 18, 20, 17, 16, 14, 13]

print("Data:", data)

min_val = data[0]
for x in data:
    if x < min_val:
        min_val = x
print("Minimum value:", min_val)

max_val = data[0]
for x in data:
    if x > max_val:
        max_val = x
print("Maximum value:", max_val)

total = 0
count = 0
for x in data:
    total += x
    count += 1
mean = total / count
print("Mean:", mean)

sorted_data = data.copy()
n = len(sorted_data)

for i in range(n):
    for j in range(0, n - i - 1):
        if sorted_data[j] > sorted_data[j + 1]:
            sorted_data[j], sorted_data[j + 1] = sorted_data[j + 1], sorted_data[j]

print("Sorted data:", sorted_data)

mid = n // 2
if n % 2 == 0:
    median = (sorted_data[mid - 1] + sorted_data[mid]) / 2
else:
    median = sorted_data[mid]

print("Median:", median)



frequency = {}

for x in data:
    if x in frequency:
        frequency[x] += 1
    else:
        frequency[x] = 1

max_freq = 0
for value in frequency.values():
    if value > max_freq:
        max_freq = value

most_frequent = []
for key, value in frequency.items():
    if value == max_freq:
        most_frequent.append(key)

print("Most frequent values:", most_frequent)
```

```python
#Task 2
import statistics

data = [12, 15, 14, 10, 18, 20, 17, 16, 14, 13]

print("Data1:", data)

print("Mean:", statistics.mean(data))

print("Median:", statistics.median(data))

std = statistics.stdev(data)
print("Standard Deviation:", std)

```
